### making function

```python
def create_invoice_product_df(dataframe, id=False):
#     For Stocks
    if id:
        return dataframe.groupby(['Invoice', "StockCode"])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
#     For Description
    else:
        return dataframe.groupby(['Invoice', 'Description'])['Quantity'].sum().unstack().fillna(0). \
            applymap(lambda x: 1 if x > 0 else 0)
```

```python
gr_inv_pro = create_invoice_product_df(df_germany)
gr_inv_pro.head()
```

```python
gr_inv_pro_df = create_invoice_product_df(df_germany, id=True)
gr_inv_pro_df.head()
```

```python
# Let's define a function to find the product name corresponding to the stock code:
df[df["StockCode"] == '10002']["Description"].unique()[0]
```

```python
# Let's define a function to find the product name corresponding to the stock code:

def check_id(dataframe, stockcode):
    product_name = dataframe[dataframe["StockCode"] == str(stockcode)]["Description"].unique()[0]
    return stockcode, product_name
```

### Determination of Association Rules

```python
frequent_itemsets = apriori(gr_inv_pro_df, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets.head()
```

###  Eclat Algorithm if we use just support insted of confidance or lift

```python
rules = association_rules(frequent_itemsets, metric="support", min_threshold=0.01)

rules.sort_values("support", ascending=False).head(5)
```

```python
rules.sort_values("lift", ascending=False).head(5)
```

# Let's explain the metrics we see in the table above:

antecedent support:

    -If X is called antecendent, 'antecedent support' computes the proportion of transactions that contain the antecedent X.
    
consequent support:

    -If Y is called consequent, 'consequent support' computes the proportion of transactions that contain the antecedent Y.
    
support:

    -'support' computes the proportion of transactions that contain the antecedent X and Y.
    
confidence:

    -Probability of buying Y when X is bought.
    
lift:

    -Represents how many times the probability of getting Y increases when X is received.

```python
sorted_rules = rules.sort_values("lift", ascending=False)
```

```python
product_id = 22554
recommendation_list = []


for idx, product in enumerate(sorted_rules["antecedents"]):
    # antecendent tuple olduğu için listeye çevirelim ve liste içinde arayalım:
    for j in list(product):
        if j == str(product_id):
            # bu yakaladığımız integer değerin indexi ne ise (idx) consequentte onu arayacağız, bulduğumuz satırlar için ilk ürünü [0]  önerelim 
            recommendation_list.append(list(sorted_rules.iloc[idx]["consequents"])[0])
            recommendation_list = list( dict.fromkeys(recommendation_list) )
```

```python
list_top5 = recommendation_list[0:5]
list_top5
```

```python
# Let's show the product names of top 5 recommended products:

for elem in list_top5:
    print(check_id(df_germany,elem))
```
